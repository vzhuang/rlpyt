```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def tail_index_estimate(data, k1=25, k2=20):
    ys = [np.sum(data[k1*x:k1*(x+1)]) for x in range(k2)]
    return np.log(k1) / (np.sum(np.log(np.abs(ys))) / k2 - np.sum(np.log(np.abs(data))) / (k1 * k2))
```

```python
# sanity check for tail index estimate
from scipy.stats import levy_stable
```

```python
a = levy_stable.rvs(1.1, 0, size=2500)
tail_index_estimate(a, 50, 50)
```

```python
def plot_hists(seed, num_iters, folder, iters_per=150):
    for i in range(num_iters):
        idx = iters_per * i
        v = np.load('log/' + folder + '/gradnoisenorms'+str(idx)+'_'+str(seed)+'.npy')
        print('Tail index estimate ', tail_index_estimate(v, 50, 40))
        plt.hist(v, bins=100)
        plt.title('Iteration ' + str(idx))
        plt.show()
```

```python
plot_hists(7742, 8, "Walker2dv2_3MGradNoise2", 100)
```
