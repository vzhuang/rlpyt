```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.load('log/gradnorms0_2377.npy')
```

```python
def plot_hists(seed, num_iters, iters_per=150):
    for i in range(num_iters):
        idx = iters_per * i
        v = np.load('log/gradnorms'+str(idx)+'_'+str(seed)+'.npy')
        plt.hist(v, bins=100)
        plt.title('Iteration ' + str(idx))
        plt.show()
```

```python
plot_hists(9649, 4)
```
